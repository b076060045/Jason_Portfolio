```python
drivepath="/Users/xuzhejia/Desktop/chrome driver/chromedriver"
browser=webdriver.Chrome(drivepath)
browser.get('https://www.nba.com/players')
time.sleep(5)
browser.find_element_by_xpath('/html/body/div[1]/div[2]/div[3]/section/div/div[2]/div[1]/div[6]/label/div/span').click()
time.sleep(8)
browser.find_element_by_xpath('/html/body/div[1]/div[2]/div[3]/section/div/div[2]/div[1]/div[7]/div/div[3]/div/label/div/select/option[1]').click()
time.sleep(5)
name = []
position = []
name_list = browser.find_elements_by_xpath('/html/body/div[1]/div[2]/div[3]/section/div/div[2]/div[2]/div/div/div/table/tbody/tr/td[1]/a/div[2]')
for x in name_list:
    name.append(x.text)
position_list = browser.find_elements_by_xpath('/html/body/div[1]/div[2]/div[3]/section/div/div[2]/div[2]/div/div/div/table/tbody/tr/td[4]')
for x in position_list:
    position.append(x.text)
```
